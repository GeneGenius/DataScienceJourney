```python
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.tail()
```

```python
sns.distplot(tips["total_bill"], kde=False, bins =50)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind ="reg")
```

```python
tips.describe()
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue="sex", palette ="magma")
```

```python
sns.rugplot(tips["total_bill"])
```

#Categorical Graphs

```python
tips.head()
```

```python
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="smoker", split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="smoker")
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="smoker")
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
sns.violinplot(x="day", y="total_bill", data=tips, )
```

```python
correlation=tips.corr()
```

```python
flights=sns.load_dataset("flights")
```

```python
flights.head()
```

```python
flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
table=flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
sns.heatmap(table)
```

```python
sns.heatmap(table, cmap="coolwarm")
```

```python
sns.heatmap(table, cmap="coolwarm", linecolor="white", linewidth=1)
```

```python
sns.clustermap(table)
```
